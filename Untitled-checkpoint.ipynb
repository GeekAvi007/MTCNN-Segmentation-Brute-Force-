```python
from PIL import Image, ImageEnhance
import torch
from facenet_pytorch import MTCNN

# Load images
image_1 = Image.open("C:\\Users\\avish\\OneDrive\\Desktop\\Segre_Model\\training.jpeg")
image_2 = Image.open("C:\\Users\\avish\\OneDrive\\Desktop\\Segre_Model\\training2.jpeg")  # Replace "path_to_your_cropped_image.jpg" with the path to your cropped image

# Preprocess images (optional)
def preprocess_image(image):
    # Example: Enhance contrast
    enhancer = ImageEnhance.Contrast(image)
    image = enhancer.enhance(1.5)  # Increase contrast by a factor of 1.5
    return image

image_1 = preprocess_image(image_1)
image_2 = preprocess_image(image_2)

# Detect if GPU cuda exists
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

# Detector MTCNN with adjusted parameters
mtcnn = MTCNN(
    select_largest=True,
    min_face_size=20,
    thresholds=[0.6, 0.7, 0.7],
    post_process=False,
    image_size=160,
    device=device
)

# Detection of bounding boxes and landmarks for both images
boxes_1, landmarks_1 = mtcnn.detect(image_1, landmarks=True)
boxes_2, landmarks_2 = mtcnn.detect(image_2, landmarks=True)

# Check if any face from image_2 is present in image_1
face_found = any(box_1[0] <= box_2[0] and box_1[1] <= box_2[1] and box_1[2] >= box_2[2] and box_1[3] >= box_2[3] for box_2 in boxes_2 for box_1 in boxes_1)

# Print result
if face_found:
    print("True: A face from image_2 is present in image_1")
else:
    print("False: No face from image_2 is present in image_1")
```
